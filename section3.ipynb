```
A=np.array([1,2,3,4])
B=np.array([100,200,300,400])
```

```
condition=np.array([True,True,False,False])
```

```
ans=[(A_val if cond else B_val) for A_val,B_val,cond in zip(A,B,condition)]
ans
```

```
ans2=np.where(condition,A,B)
ans2
```
